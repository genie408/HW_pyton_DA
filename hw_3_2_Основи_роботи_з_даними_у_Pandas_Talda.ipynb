Важливо навчитись висловлювати свої гіпотези щодо даних, і з часом та досвідом вони будуть ставати щоразу глибше і точніше. Весь аналіз даних будується на тому, що ми перевіряємо певні гіпотези та робимо висновки і рекоемндації на основі своїх досліджень.

```python
filt_Yan = df [df.City == 'Yangon']
filt_Man = df [df.City == 'Mandalay']
```

```python
mean_gr_in_Y = filt_Yan ['gross income'].mean()
mean_gr_in_M = filt_Man ['gross income'].mean()
display (mean_gr_in_Y, mean_gr_in_M)
```

Як бачимо середнє значення gross income у місті Yangon дещо менший ніж у місті Mandalay, хоча кількість інвойсів у Yangon незначно але більша ніж у Mandalay.
gross income в одному місті може бути більшим ніж у іншому через:
#### 1 більшу загальну сумму продажів
#### 2 тому що клієнти більше купують певну групу товарів
#### 3 тому що клієнти надають перевагу певному способу оплати
Але в нашому випадку...

```python
df ['tax_eq_gi'] = df ['Tax 5%'] == df ['gross income']
df ['tax_eq_gi'].value_counts()
```

Отже в нашому випадку gross income дорівнює значенню 5% податку, а значить залежить лише від загальної суми продажів. Чим більшою буде сума продажів тим більшим буде gross income.
